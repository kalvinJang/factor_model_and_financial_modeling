```python
시가총액_6_KOSPI = pd.DataFrame()
for i in month6:
    temp = pd.read_csv(i,encoding = 'cp949')
    시가총액_6_KOSPI_temp = pd.DataFrame(temp[temp['시장구분']=='KOSPI']['시가총액'])
    시가총액_6_KOSPI_temp.index=temp[temp['시장구분']=='KOSPI']['종목코드'].apply(lambda x: str(x).rjust(6, '0'))
    시가총액_6_KOSPI_temp.columns = [i[:-4].replace('_','/')]
    시가총액_6_KOSPI = pd.concat([시가총액_6_KOSPI, 시가총액_6_KOSPI_temp], axis=1, join='outer')
시가총액_6_KOSPI = 시가총액_6_KOSPI.T
시가총액_6_KOSPI   #KOSPI는 1770개,  KOSDAQ은 2431개
```

```python
시가총액_6_KOSPImed = 시가총액_6_KOSPI.median(axis=1)
```

```python
시가총액_6 = pd.DataFrame()
for i in month6:
    temp = pd.read_csv(i,encoding = 'cp949')
    시가총액_6_temp = pd.DataFrame(temp['시가총액'])
    시가총액_6_temp.index=temp['종목코드'].apply(lambda x: str(x).rjust(6, '0'))
    시가총액_6_temp.columns = [i[:-4].replace('_','/')]
    시가총액_6 = pd.concat([시가총액_6, 시가총액_6_temp], axis=1, join='outer')
시가총액_6 = 시가총액_6.T
시가총액_6  #전체는 4349개  #KOSPI는 1770개
```

```python
시가총액_6_KOSPImed
```

```python
big_f = 시가총액_6.apply(lambda x: x>시가총액_6_KOSPImed, axis=0)
small_f = 시가총액_6.apply(lambda x: x<=시가총액_6_KOSPImed, axis=0)
```

```python
def get_true(row):
    return list(row[row==True].index)
```

```python
# 6월 말 기준, Big과 Small 종목코드
SMB_stocks = pd.concat([big_f.apply(lambda x: get_true(x), axis=1), small_f.apply(lambda x: get_true(x), axis=1)], axis=1)
SMB_stocks.columns = ['Big', 'Small']
SMB_stocks.index = pd.to_datetime(SMB_stocks.index, format='%y/%m/%d')
SMB_stocks
```

```python
SMB_stocks = SMB_stocks.resample('YS', kind='Timestamp').first()
SMB_stocks
```

```python
daily등락률 = pd.read_csv('./../KR data daily등락률.csv')
daily등락률.index = daily등락률['종목코드'].apply(lambda x: str(x).rjust(6,'0'))
daily등락률 = daily등락률.iloc[:, 1:].T
daily등락률.index = pd.to_datetime(daily등락률.index, format="%y/%m/%d")
daily등락률 = (daily등락률/100)+1
daily등락률
```

```python
r_win = daily등락률.apply(lambda x: winsorize(x, limits=[0.01, 0.01]), axis=0)
monthly등락률 = r_win.resample('M').prod()
monthly등락률
```

```python
#daily등락률 확인
daily등락률.groupby([daily등락률.index.year, daily등락률.index.month]).get_group((1995, 5)).prod(axis=0)
```

#### price, 시가총액, 장부가치(자본총계-우선주 자본금), 유형자산변화분, 재고자산 변화분, 전년도 총자산, 당기순이익(또는 경상이익), 전년도 장부가치


- 시장수익률: KOSPI+KOSDAQ 시총가중 수익률  (1996년 7월 1일 이후에 코스닥종목 편입)
- 무위험이자율: CD91일물 수익률(1993.3 ~), 콜금리 또는 정기예금금리(1983. 4 ~ 1991.2)
- SMB : 시가총액 (우선주 고려 안 함 // 코스피 시총의 median이 breakpoint) 6월 말 기준
    - 시가총액은 KOSPI데이터 중 6월 마지막 거래일에 맞춰서 '종목코드', '시가총액' column만 받고, 종목코드를 index로 지정해서 concat(outer)하기
    - KOSPI종목 시가총액 median 구해서 전체시장에서 big firm, small firm 종목코드를  나누기

- **HML :** 자본총계 / 시장가치(시가총액) 전년도 12월 말기준(시총은 6월, 12월 두 개 구해야함) // breakpoint는 코스피 시총의 30%, 70%이며 코스닥 종목들은 범위에 맞춰서 넣는다. (따라서 코스닥 종목들은 개수로 자르는 게 아니라, 시총 범위에 맞춰서 넣음)
- **RMW** =  전년도12월말 (영업손익 - 이자비용or금융비용) / 전년도12월말 총자산
- **CMA =** ROA : 전년도12월말 당기순이익 / t-2년도 12월말 총자산
