### 분위 정하기 (SMB은 mktcap기준 1:3 // value, growth는 HML은 BE/ME기준 3:4:3)
#### FF논문에 Size의 기준은 NYSE의 median으로 잡았고, 그러면 비율이 1:3정도 됐다고 함)
#### 6월 30일 기준으로 ranking 정해서 분위 정함

```python
mktcap_rank = mktcap.rank(1, 'first').apply(lambda x: pd.qcut(x, q=4, labels = range(1,5)) if not x.isnull().all() else x, 1)
mktcap_6m = mktcap_rank.groupby(mktcap_rank.index.month).get_group(6)
```

```python
def extract_portfolio(data, n): #여기서 data는 각 date에 해당하는 series, (series의 index는 티커, name은 date)
    extract = pd.DataFrame(data==n)
    return list(extract[extract[data.name]==True].index)
```

```python
def portfolio_groups(data, r):
    return [extract_portfolio(data, n) for n in r]
```

```python
mktcap_port = mktcap_6m.apply(lambda x: portfolio_groups(x, range(1,5)), axis=1,result_type = 'expand')
mktcap_port.columns = ['MktCap{}'.format(i) for i in range(1,5)]
mktcap_port
```

```python
BVMV = bv/mktcap
BVMV
```

```python
HML = bv/mktcap
HML = HML.groupby(HML.index.month).get_group(6)
```

```python
HML_q = HML.rank(1, 'first').apply(lambda x: pd.qcut(x, q=10, labels = range(1,11)) if not x.isnull().all() else x, 1)
HML_port = HML_q.apply(lambda x: portfolio_groups(x, range(1,11)), axis=1, result_type = 'expand')
HML_port.columns = ['HML{}'.format(i) for i in range(1,11)]
HML_port
```

### 2 X 3으로 만들기

```python
def makethree(db, position):
    return [db[db.columns[position]] + db[db.columns[position+1]]+db[db.columns[position+2]]]
```

```python
def makefour(db, position):
    return [db[db.columns[position]] + db[db.columns[position+1]]+db[db.columns[position+2]]+db[db.columns[position+3]]]
```

```python
SMB_2 = pd.concat([mktcap_port['MktCap1'], pd.DataFrame(makethree(mktcap_port, 1)).T], axis=1)
SMB_2.columns = ['Big', 'Small']
SMB_2
```

```python
HML_3 = pd.concat([pd.DataFrame(makethree(HML_port,0)).T, pd.DataFrame(makefour(HML_port, 3)).T, pd.DataFrame(makethree(HML_port,7)).T], axis=1)
HML_3.columns = ['High', 'Medium', 'Low']
HML_3
```

### 월별 price데이터가 없는 관계로.... 수익률은 mktcap으로 계산
먼저 windorize해서 위아래 5% 아웃라이어 없애기     
매월마다 팩터별 equally weighted 포트폴리오의 수익률을 계산 후 1/3*(small-big) 1/2*(high-low)한다

```python
SMB_2_12m = pd.DataFrame(np.repeat(SMB_2.values, 12, axis=0), columns=SMB_2.columns)[:-5]
SMB_2_12m.index = mktcap.loc['1987-06-30':,:].index
SMB_2_12m
```

```python
mc_pct = mktcap.pct_change()
mc_pct
```

```python
SMB_r = pd.DataFrame(index=SMB_2_12m.index, columns=SMB_2.columns)
r_big = []
r_small = []
for date in SMB_2_12m.index:
    r_big.append(mc_pct.loc[date, SMB_2_12m.loc[date, 'Big']].mean())
    r_small.append(mc_pct.loc[date, SMB_2_12m.loc[date, 'Small']].mean())
SMB_r['Big']= r_big
SMB_r['Small'] = r_small
SMB_r
```

```python
SMB_r['Diff'] = 1/3*(SMB_r['Big'] - SMB_r['Small'])
```

```python
HML_3_12m = pd.DataFrame(np.repeat(HML_3.values, 12, axis=0), columns=HML_3.columns)[:-5]
HML_3_12m.index = mktcap.loc['1987-06-30':,:].index
HML_3_12m
```

```python
HML_r = pd.DataFrame(index=HML_3_12m.index, columns=['High', 'Low'])
r_high = []
r_low = []
for date in HML_3_12m.index:
    r_high.append(mc_pct.loc[date, HML_3_12m.loc[date, 'High']].mean())
    r_low.append(mc_pct.loc[date, HML_3_12m.loc[date, 'Low']].mean())
HML_r['High']= r_high
HML_r['Low'] = r_low
HML_r['Diff'] = 1/2*(HML_r['High'] - HML_r['Low'])
HML_r
```

```python
Factors = pd.concat([SMB_r['Diff'], HML_r['Diff']], axis=1, join='inner')
Factors.columns = ['SMB', 'HML']
Factors = Factors*100
```

```python
Factors.to_csv('3Factor.csv')
```
