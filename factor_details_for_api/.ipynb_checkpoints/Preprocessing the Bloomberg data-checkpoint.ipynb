```python
def shift_q(date):
        if date[4:6]=='01' or date[4:6]=='02'or date[4:6]=='03' :
            date = date[:4] + '0331'
        elif date[4:6]=='04' or date[4:6]=='05'or date[4:6]=='06' :
            date = date[:4] + '0630'
        elif date[4:6]=='07' or date[4:6]=='08' or date[4:6]=='09' :
            date = date[:4] + '0930'
        elif date[4:6]=='10' or date[4:6]=='11'or date[4:6]=='12' :
            date = date[:4] + '1231'
        return date
```

```python
kospi = pd.read_csv('kospi ff5 data.csv', encoding='cp949', header = 1, index_col=0).iloc[1:,:]
kospi
```

```python
def preprocessing_bloomberg(data, number_of_series, *args):
    temp = data.index.map(lambda x: x[:4]+x[5:7]+x[8:10])
    kospi.index = list(map(lambda x: shift_q(x), temp))
    preprocess = dict()
    for i in range(number_of_series):
        finance = kospi.iloc[:,i::number_of_series]
        finance.columns = list(map(lambda x: x[:6], finance.columns))
        real = pd.DataFrame()
        for j in list(set(finance.index)):
            temp = finance.loc[[j],:]
            if temp.shape[0]>1:
                temp.reset_index(inplace=True)
                while temp.shape[0]!=1:
                    temp = temp.append(pd.merge(temp.iloc[[0],:],temp.iloc[[1],:], 'outer'))
                    temp = temp.iloc[:,1:].reset_index().iloc[:, 1:]
                    temp = temp.drop([0,1])
                temp.index= [j]
                real = real.append(temp)
            else:
                real = real.append(temp)
        preprocess[str(arg[i])]=real
    return preprocess
```
